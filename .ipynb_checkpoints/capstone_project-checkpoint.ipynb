# The Impact of The Demographical Structure of the U.S. on Migration
### Data Engineering Capstone Project

#### Project Summary
<p>The U.S. government is planning on a new immigration policy. To make a better decision, it kicked off a new initiative with its data collection department. The purpose is to get more insight on the impact of demographical structure on the profile of incomers as well as the volume and the pace of migration across the <b>states</b>.</p>

<p>For this reason, the department has collected the immigration data from the National Tourism and Trade Office where respondents' details are kept in record. Additionally, U.S. City Demographic data from 2015 has been used to find the relationship between immigration and demography of the U.S. The data will also be used to predict the profile of incomers arriving to the U.S. and make new pathways and policies in advance.<p>

<p>Since this is a pilot program, the team has been assigned to work on the Immigration Data only for January and February of 2016. Considering the data will continue to grow, and be demanded from different State Deparments, there needs to be a process of providing clean data in a scalable platform. The rationale of the architecture will be provided in the next section.</p>

### 1. Scope of the Project and Data Gathering

#### 1.1. Data Gathering
Two data sets will be used to make this project realize. One of them is the immigration data (https://travel.trade.gov/research/reports/i94/historical/2016.html) and the demographical data (https://public.opendatasoft.com/explore/dataset/us-cities-demographics/export/). The immigration data includes traveler's official issued document numbers, flight details, visa status, and their background. The demographics data includes statistical information of each state based on race, resident's status (foreign, veteran), and gender.

#### 1.2. Scope 
The scope of the project includes cleaning, extracting, transforming and loading the immigration and demographical data to a relational database using star schema model. With the end product, data analysists and data scientists will be able to analyse how the migration in January and February 2016 got affected by the demographical structure of the U.S. in 2015.
##### 1.2.1. Cleaning
Both immigration and demography data come from different sources and land on AWS's Simple Storage Service (S3). For both the data to be used properly for analysis and prediction, they need to be cleaned from duplicates and null values. Moreover, not all the fields in the provided data will be used for the consuming departments. Therefore, necessary drop of fields will occur in the data before taking any further action. 
##### 1.2.2. Extracting
The immigration data is in a sas7bdat format and relatively large as each month of data is approximately 4 million rows 28 columns. Hence, a distributed processing tool, Spark, has been used to handle such large volume of data. Demographical data comes with 12 columns and has approximately 3000 rows as a csv format. 
##### 1.2.3. Transforming
In the tranformation process, necessary modifications have been applied to the data, to make it clean and easily understable for the consuming groups. For immigration data, this process varies from changing data types of the columns to modifying date columns using Spark's user defined functions. With regards to demographical data, a simple aggregation and joining have been applied to make the data ready for loading.
##### 1.2.4. Loading
After the transformation, the data is stored in S3 as staging. A star schema model has been built to load the data from staging to a datawarehouse. Star schema is commonly used for modelling relational data warehouses. The SQL queries will be much easier to run thanks to denormalization of the data within dimension tables. As a datawarehouse, AWS' datawarehouse product Redshift is used in this project. The reason for picking a datawarehouse is to ease the load of access to database since the data will be used from multiple departments.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import from_unixtime, unix_timestamp, col, round, udf
import datetime
spark = SparkSession.builder.\
config("spark.jars.packages","saurfang:spark-sas7bdat:2.0.0-s_2.11")\
.enableHiveSupport().getOrCreate()
df_spark =spark.read.format('com.github.saurfang.sas.spark').load('../../data/18-83510-I94-Data-2016/i94_jan16_sub.sas7bdat')
```

```python
df_spark.show(10)
```

```python
df_spark = df_spark.withColumn("month1", col("i94mon").cast("integer"))
df_spark = df_spark.withColumn("month2", col("month1").cast("string"))
df_spark = df_spark.withColumn("month", from_unixtime(unix_timestamp(col("month2"),'M'),'MM'))
df_spark = df_spark.withColumn("immi_key", col("cicid").cast("integer"))
df_spark = df_spark.withColumn("year1", col("i94yr").cast("integer"))
df_spark = df_spark.withColumn("year2", col("year1").cast("string"))
df_spark = df_spark.withColumn("year", from_unixtime(unix_timestamp(col("year2"),'YYYY'),'YYYY'))
df_spark = df_spark.drop("i94mon","month1","month2","cicid","i94yr","year1","year2")
df_spark = df_spark.withColumn("cit1", col("i94cit").cast("integer"))
df_spark = df_spark.withColumn("cit", col("cit1").cast("string"))
df_spark = df_spark.withColumn("res1", col("i94res").cast("integer"))
df_spark = df_spark.withColumn("res", col("res1").cast("string"))
df_spark = df_spark.drop("i94cit","cit1","i94res","res1")
df_spark = df_spark.withColumn("visa1", col("i94visa").cast("integer"))
df_spark = df_spark.withColumn("visa_type", col("visa1").cast("string"))
df_spark = df_spark.withColumn("age_respondent", col("i94bir").cast("integer"))
df_spark = df_spark.withColumn("biryear1", col("biryear").cast("integer"))
df_spark = df_spark.withColumn("biryear2", col("biryear1").cast("string"))
df_spark = df_spark.withColumn("birth_year", from_unixtime(unix_timestamp(col("biryear2"),'YYYY'),'YYYY'))
df_spark = df_spark.drop("i94visa","visa1","i94bir","biryear","biryear1","biryear2")
df_spark = df_spark.drop("count")
df_spark = df_spark.withColumn("arr_date1", col("arrdate").cast("integer"))
df_spark = df_spark.withColumn("dept_date1", col("depdate").cast("integer"))
```

```python
def date_time_conversion(val):
    epoch = datetime.datetime(1960, 1, 1)
    try:
        return (epoch + datetime.timedelta(days=val)).date().strftime("%Y-%m-%d")
    except ValueError:
        return None
    except TypeError:
        return None
    
date_convert_udf = udf(date_time_conversion)
```

```python
df_spark = df_spark.withColumn("arrival_date", date_convert_udf(col("arr_date1")))
df_spark = df_spark.withColumn("departure_date", date_convert_udf(col("dept_date1")))
df_spark = df_spark.drop("arrdate","arr_date1","depdate","dept_date1")
df_spark = df_spark.withColumn("arrival_type1", col("i94mode").cast("integer"))
df_spark = df_spark.withColumn("arrival_type", col("arrival_type1").cast("string"))
df_spark = df_spark.drop("arrival_type1","i94mode")
```

```python
def dtad_format(val):
    try:
        return datetime.datetime.strptime(val, '%Y%m%d').date().strftime("%Y-%m-%d")
    except ValueError:
        return None
    except TypeError:
        return None
    
dtad_udf = udf(dtad_format)
```

```python
df_spark = df_spark.withColumn("date_form_added", dtad_udf(col("dtadfile")))
df_spark = df_spark.na.drop(subset=["date_form_added"])
df_spark = df_spark.drop("dtadfile")
df_spark = df_spark.drop("insnum")
```

```python
def dtaddto_format(val):
    try:
        return datetime.datetime.strptime(val, '%m%d%Y').date().strftime("%Y-%m-%d")
    except ValueError:
        return None
    except TypeError:
        return None
dtaddto_format_udf = udf(dtaddto_format)
```

```python
df_spark = df_spark.withColumn("admission_date", dtaddto_format_udf(col("dtaddto")))
df_spark = df_spark.drop("dtaddto")
df_spark = df_spark.na.drop(subset=["visapost"])
df_spark = df_spark.drop("entdepu")
df_spark = df_spark.na.drop(subset=["entdepd"])
df_spark = df_spark.drop_duplicates(subset=['admnum'])
df_spark = df_spark.drop("occup")
df_spark = df_spark.drop("immi_key")
```

```python
df_spark = df_spark.drop_duplicates(subset=['admission_date','i94port','i94addr','matflag','gender','airline','fltno','visatype','month',
                                            'year','cit','res','visa_type','age_respondent','birth_year','arrival_date','departure_date','arrival_type','admission_date'])
df_spark = df_spark.na.drop(subset=['admission_date','i94port','i94addr','matflag','gender','airline','fltno','visatype','month',
                                    'year','cit','res','visa_type','age_respondent','birth_year','arrival_date','departure_date','arrival_type','admission_date'])
```

### 3. Define the Data Model
#### 3.1 Conceptual Data Model
Star schema model has been chosen to build the pipeline in a Redshift datawarehouse. Staging tables arrive in the datawarehouse after being cleaned and transformed. These two tables are copied from S3 bucket and loaded into Redshift. After the initial staging load, the two tables are processed to fit into the star schema model. From the immigration staging table, the data is split into 2 dimensional tables. The reason is to seperate respondents' sensitive information from generic information. For example, immi_file_dim table could hold the information such as the reason for coming to the U.S.(arrival_type), where they landed, what their admission date was and so on whereas respondent_dim_table holds information about respondent's background information such as where they came from and so on. demographic_dim_table holds the statistical information about the U.S. states. With this approach we can easily do analysis on demographical structure of the states and their relation to the profile of the respondent. For example, we can find if incomers' choice of place of arrival could be linked to the difference between race populations in each state. Fact table holds the primary keys of each dimension table which makes it easy to join tables.

#### 5. Complete Project Write Up
<p>Spark is used to perform Extract Transform Load process. This tools was used because Spark can compute processing in a distributed machines of clusters. The cleaning and ETL process were written in Python. Python is vastly used because it has many convenient libraries to handle data engineers task. For example, we connect to Redshift using psycopg2 library. I chose Redshift datawarehouse because it can handle heavy amount of queries being run simultanouesly. This way, multiple departments can access the data more efficiently.</p>

<p>The data should be further processed for the coming months of immigration data. The project is based on analysing the immigration data on a monthly basis. So once a month will suffice for analysis.</p>
<p>The hypothetical scenario: <br>
<p>1- The data was increased by 100x.</p>
<p> Storage wise, it would be beneficial to choose cloud storage services such as AWS S3 storage. Cloud storage systems are cheap and easy to maintain compared to maintaning the storage by yourself. Additionally, we can classify data based on the frequency of access. This way, we don't need to pay more for the data that's no longer in use (or won't be used for a long time.). We can also partition the data by year, month, day and hour. This way, we can adjust our job processing frequency to a more frequent run.</p>
